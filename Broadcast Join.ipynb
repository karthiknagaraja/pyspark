```
from pyspark.sql.functions import broadcast
```

```
df1 = spark.createDataFrame([('a',2),('b',3),('c',4),('c',44)],['A','B'])
```

```
df1.show()
```

```
df2 = spark.createDataFrame([('a','aaaa'),('b','bbbb'),('c','cccc')],('A','C'))
```

```
df1.join(df2, df1.A == df2.A,'inner').show()
```

```
df1.join(broadcast(df2),'A').show()
```
