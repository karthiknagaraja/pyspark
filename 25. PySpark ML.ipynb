```
from pyspark.ml.classification import LogisticRegression
```

```
from pyspark.ml.feature import VectorAssembler
```

```
income_df.dtypes
```

```
filter(lambda x: not x.endswith("_tf"), income_df.columns)
```

```
vec_cols = map(lambda x:x[0], filter(lambda c: c[1] == 'int' or c[1] =='vector'  ,income_df.dtypes))
vec_cols.remove('income_tf_en')
```

```
vec_cols = map(lambda x:x[0], filter(lambda c: c[1] =='vector'  ,income_df.dtypes))
vec_cols.remove('income_tf_en')
vec_cols.extend(['age',
 'fnlwgt',
 'educational-num',
 'capital-gain',
 'capital-loss',
 'hours-per-week'])
```

```
ve = VectorAssembler(inputCols=vec_cols, outputCol='vec_out_features')
```

```
income_df = ve.transform(income_df)
```

```
display(income_df)
```

```
train ,test = income_df.randomSplit([.7,.3])
```

```
from pyspark.ml.classification import LogisticRegression
```

```
lr = LogisticRegression(regParam=0.001, labelCol='income_tf', featuresCol='vec_out_features')
```

```
model = lr.fit(train)
```

```
output = model.transform(test)
```

```
display(output)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
bce = BinaryClassificationEvaluator(labelCol='income_tf')
```

```
bce.evaluate(output)
```

```
stringindexers = []
for col in cat_cols:
  stringindexers.append(StringIndexer(inputCol=col, outputCol=col+'_tf'))
```

```
stages = []
stages.extend(stringindexers)
stages.append(ve)
stages.append(lr)
```
